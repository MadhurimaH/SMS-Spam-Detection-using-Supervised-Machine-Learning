```python
# This code calculates the occuarance of continous dots/ellipses in a message
import re
def getDot(str): 
    c=0
    dots = re.findall('\.{2,}', str)
    if dots:
        c=c+len(dots)
       # print(dots)
    
    return c  


import pandas as pd
data=pd.read_csv(r"C:\Users\Vasvi\Desktop\tanya_hp\tanya_new\bi\spam_f.csv",encoding="ISO-8859-1")

data['Dot']= data['Message'].apply(getDot)

data
```

```python
#This code generates whether ellipses/dots are present in the message, it uses 0 to mark their absence and 1 for presence

import re
# Function to extract all the numbers from the given string 
def getDot(str): 
    dots = re.findall('\.{2,}', str)
    if dots:
        c=1
    else:
        c=0
    return c  


import pandas as pd
data=pd.read_csv(r"C:\Users\Vasvi\Desktop\tanya_hp\tanya_new\bi\spam_f.csv",encoding="ISO-8859-1")

data['Dot']= data['Message'].apply(getDot)
data

```
