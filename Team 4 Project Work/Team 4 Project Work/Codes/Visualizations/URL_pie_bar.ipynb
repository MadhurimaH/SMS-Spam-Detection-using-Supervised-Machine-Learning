```python
#This notebook containes data visualisation performed on URL attribute. It creates a pie chart and bar graph for the same.

from urlextract import URLExtract
extractor = URLExtract()
def getURL(str): 
    urls = extractor.find_urls(str)
    if urls:
        c=1
       # print(str)
    else:
        c=0
       # print(str)
    return c    
     
import pandas as pd
data=pd.read_csv(r"C:\Users\Vasvi\Desktop\tanya_hp\tanya_new\bi\spam_f.csv",encoding="ISO-8859-1")

data['URL']= data['Message'].apply(getURL)
data
```

```python
df=data.loc[data['URL'] == 1]
x=[]
x.append(df.loc[df['Category']=="ham"].shape[0])
x.append(df.loc[df['Category']=="spam"].shape[0])
print(x)
```

```python
import matplotlib.pyplot as plt
y= ["Ham","Spam"]
colors = ["#2E8B57", "#CD5C5C"]
explode = (0.06, 0)
plt.pie(x, labels=y, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Percentage of Spam vs. Ham for messages where a URL is present")
#plt.savefig(,300)
plt.show()
```

```python
#objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
import numpy as np
y_pos = np.arange(len(y))

fig, ax = plt.subplots()
b=plt.bar(y_pos, x,  align='center',alpha=0.7)
b[0].set_color("#2E8B57")
b[1].set_color("#CD5C5C")
#bar_width = 0.35
plt.xticks(y_pos, y)
#plt.fit()
plt.ylabel('No. of messages')
plt.title('No. of Spam and Ham messages for texts with URLs')
plt.tight_layout()
```
