```python
import re
import pandas as pd
data=pd.read_csv(r"kaggle.csv",encoding="ISO-8859-1")
# Function to extract all the numbers from the given string 
def getEmo(str): 
    c=0
    smileys = """:-) :) :o) :] :3 :c) :> =] 8) =) :} :^) 
             :D 8-D 8D x-D xD X-D XD =-D =D =-3 =3 B^D""".split()
    pattern2 = "|".join(map(re.escape, smileys))
    emo=re.findall(pattern2, str)
    if emo:
        c=len(emo)
    else:
        c=0
    return c  


data['Emoji']= data['Message'].apply(getEmo)


#function to extract mathematical symbols from a given string
def mathsymbols(str):
    mathsym = 0
    for i in range(0, len(str)):  
            ch = str[i]  

            if (ch=='+' or ch == '-' or ch=='*' or ch=='^' or ch=='%' or ch=='/' or ch=='=' ):    
                mathsym += 1
             
    return mathsym
data['Mathemtical symbols']= data['Message'].apply(mathsymbols)
#avgspam=0.0;
#avgham=0.0;
from urlextract import URLExtract
extractor = URLExtract()
def getURL(str): 
    urls = extractor.find_urls(str)
    if urls:
        c=len(urls)
    else:
        c=0
       # print(str)
    return c    
     
data['URL']= data['Message'].apply(getURL)
data




```

```python
import csv
import re
#function to get wether a msg contain shorthand or not
def translator(user_string):
    user_string = user_string.split(" ")
    j = 0
    c=0
    for _str in user_string:
        # File path which consists of Abbreviations.
        fileName = "slang.txt"
        # File Access mode [Read Mode]
        accessMode = "r"
        with open(fileName, accessMode) as myCSVfile:
            # Reading file as CSV with delimiter as "=", so that abbreviation are stored in row[0] and phrases in row[1]
            dataFromFile = csv.reader(myCSVfile, delimiter="=")
            # Removing Special Characters.
            _str = re.sub('[^a-zA-Z0-9-_.]', '', _str)
            for row in dataFromFile:
                # Check if selected word matches short forms[LHS] in text file.
                if _str.upper() == row[0]:
                    c=c+1
                    # If match found replace it with its appropriate phrase in text file.
                #return c
            myCSVfile.close()
        j = j + 1
    return c

data['shorthands']= data['Message'].apply(translator)
data.head()
```
