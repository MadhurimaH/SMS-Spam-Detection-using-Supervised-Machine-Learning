```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"D:\desktop material\igd\MCA 4\bi\project files\spam2.csv",encoding="ISO-8859-1")

import matplotlib.pyplot as plt
y= ["Ham","Spam"]

list1=data.mis_spell.unique()
print(list1)
barwidth=0.3;
for i in range(len(list1)):
    df=data.loc[data['mis_spell'] ==list1[i]]
    k=[]
    k.append(df.loc[df['Category']=="ham"].shape[0])
    k.append(df.loc[df['Category']=="spam"].shape[0])
    
    y_pos = np.arange(len(y))

    fig, ax = plt.subplots()
    b=plt.bar(y_pos+0.35, k,barwidth,align='center',alpha=0.7)
    b[0].set_color("#2E8B57")
    b[1].set_color("#CD5C5C")
    #bar_width = 0.35
    plt.xticks(y_pos, y)
    #plt.fit()
plt.ylabel('No. of messages')
plt.title('No. of ')
plt.tight_layout()
plt.show();
```
