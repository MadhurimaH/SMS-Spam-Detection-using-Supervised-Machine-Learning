```python
import pandas as pd
#load dataset
ds=pd.read_csv("dataset/SPAM.csv")
ds['upper'] = ds['Message'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
length=len(ds)
i=0
c_up_ham=[]
c_up_spam=[]
for index, row in ds.iterrows():
    if(row['Category']=="ham"):
       c_up_ham.append(row['upper'])
    else :
        c_up_spam.append(row['upper'])
```

```python
ds['word_count'] = ds['Message'].apply(lambda x: len(str(x).split(" ")))
count_ham=[]
count_spam=[]
for index, row in ds.iterrows():
    if(row['Category']=="ham"):
        count_ham.append(row['word_count'])
    else:
        count_spam.append(row['word_count'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.scatter(count_ham, c_up_ham,c='g')
plt.scatter(count_spam,c_up_spam,c='r')
plt.xlabel('count of words')
plt.ylabel('no of capitals')

plt.show()
```
